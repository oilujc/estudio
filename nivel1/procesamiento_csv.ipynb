```python
# Importar la libreria csv para poder leer y procesar archivos CSV

import csv

datos = csv.reader(open('../online_sales_dataset.csv', 'r', encoding='utf-8'))
```

```python
# Calcular el total de ventas por cada producto
productos = {}

for venta in list(datos)[1:]:
    # StockCode: 1, Description: 2, Quantity: 3, UnitPrice: 5, Discount: 8, ShippingCost: 10

    stock_code = venta[1]
    description = venta[2]
    quantity = int(venta[3])
    unit_price = float(venta[5])
    discount = float(venta[8])
    shipping_cost = float(venta[10]) if venta[10] else 0.0

    total_venta = (unit_price * quantity) - discount + shipping_cost


    if stock_code not in productos:
        productos[stock_code] = {
            'numero_ventas': 1,
            'nombre': description,
            'cantidad': quantity,
            'total_venta': total_venta,
        }
    else:
        productos[stock_code]['cantidad'] += quantity
        productos[stock_code]['total_venta'] += total_venta
        productos[stock_code]['numero_ventas'] += 1
 
```

```python
for sku, datos_producto in productos.items():
    print(f"SKU: {sku}, Nombre: {datos_producto['nombre']}, "
          f"Total productos vendidos: {datos_producto['cantidad']}, "
          f"Total venta: {datos_producto['total_venta']:.2f}, "
          f"Número de ventas: {datos_producto['numero_ventas']}")
```

```python
with open('../resultado_ventas.csv', 'w', encoding='utf-8') as csvfile:
    
    csvfile.write('SKU,Nombre,Total productos vendidos,Total venta,Número de ventas\n')
    for sku, datos_producto in productos.items():
        csvfile.write(f"{sku},{datos_producto['nombre']},{datos_producto['cantidad']},"
                      f"{datos_producto['total_venta']:.2f},{datos_producto['numero_ventas']}\n")
       
```
